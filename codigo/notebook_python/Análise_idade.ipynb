**Importação dos dados**

```python
import pandas as pd
microdados_limpo_csv = pd.read_csv(r'microdados_limpo.csv', sep=';')
```

**Ver dados**

```python
microdados_limpo_csv
```

**Gráfico**

```python
# Import plotly and create a figure
import plotly.graph_objects as go
fig = go.Figure()


# Add the histogram traces to the figure
for column_header in ['idade']:
    fig.add_trace(go.Histogram(x=microdados_limpo_csv[column_header], name=column_header))

# Update the layout
# See Plotly documentation for customizations: https://plotly.com/python/reference/histogram/
fig.update_layout(
	xaxis_title='idade',
    title='idade frequencies',
    barmode='group'
)
fig.show(renderer="iframe")
```
